```python
from sklearn.datasets import load_iris
```

```python
#fit the dataset into data frame
iris_data=load_iris()
```

```python
print(iris_data)
```

```python
#split the data into features and target data
x=iris_data.data
y=iris_data.target
```

```python
x
```

```python
y
```

```python
#use the KFOLD Technique to split the data
from sklearn.model_selection import KFold
```

```python
kfold=KFold(n_splits=5, random_state=None) 
```

```python
#we split the x data into  10 kfolds using a for loop
print('train_set:     test_set      ')
for train_set,test_set in kfold.split(x):
    print(train_set,test_set)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_class=RandomForestClassifier(n_estimators=10)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score=cross_val_score(rf_class,x,y,scoring='accuracy', cv=10)
```

```python
cv_score
```

```python
accuracy=cv_score.mean()*100
```

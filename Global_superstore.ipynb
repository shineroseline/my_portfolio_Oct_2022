```python
import re
p = re.compile('$()')
new_superstore_data['Sales'] = [p.sub('', x) for x in new_superstore_data['Sales'].tolist()]
```

```python
new_superstore_data.head()
```

```python
spec_chars = ['$',"(", ")"]
for char in spec_chars:
    new_superstore_data['Sales'] = new_superstore_data['Sales'].str.replace(char, ' ')
```

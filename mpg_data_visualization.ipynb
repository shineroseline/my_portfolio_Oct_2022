```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler(copy=True, feature_range=(0, 1))
```

```python
data_scale=auto_data.copy()
```

```python
data_scale['displacement'] = scale.fit_transform(auto_data['displacement'].values.reshape(-1, 1))
#data_scale['horsepower'] = scale.fit_transform(data['horsepower'].astype(float).values.reshape(-1, 1))
data_scale['acceleration'] = scale.fit_transform(auto_data['acceleration'].values.reshape(-1, 1))
data_scale['weight'] = scale.fit_transform(auto_data['weight'].values.reshape(-1, 1))
data_scale['mpg'] =scale.fit_transform(auto_data['mpg'].values.reshape(-1, 1))
```

```python
data_scale.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#using a boxplot, we show the relationship between mpg and country
style.use('ggplot')
data_scale_plot = pd.concat([data_scale['mpg'], data_scale['Country_code']], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sb.boxplot(x='Country_code', y="mpg", data=data_scale_plot)
fig.axis(ymin=0, ymax=1)
plt.axhline(data_scale.mpg.mean(),color='r',linestyle='dashed',linewidth=2)
#we note that cars from USA as mpg lower than the global average while cars from EUROPE AND JAPAN has mpg greater than the global average.
```

```python
style.use('ggplot')
data_scale_plot = pd.concat([data_scale['mpg'], data_scale['model_year']], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sb.boxplot(x='model_year', y="mpg", data=data_scale_plot)
fig.axis(ymin=0, ymax=1)
plt.axhline(data_scale.mpg.mean(),color='r',linestyle='dashed',linewidth=2)
```

```python
style.use('ggplot')
data_scale_plot = pd.concat([data_scale['mpg'], data_scale['cylinders']], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sb.boxplot(x='cylinders', y="mpg", data=data_scale_plot)
fig.axis(ymin=0, ymax=1)
plt.axhline(data_scale.mpg.mean(),color='r',linestyle='dashed',linewidth=2)
```

```python
#we check the correlation among the features
corrmat = data_scale.corr()
f, ax = plt.subplots(figsize=(12, 9))
sb.heatmap(corrmat, square=True)
```

```python
#we check correlation among some selected features
factors = ['cylinders','displacement','horsepower','acceleration','weight','mpg']
corrmat = data_scale[factors].corr()
f, ax = plt.subplots(figsize=(12, 9))
sb.heatmap(corrmat, square=True)
```

```python
import numpy as np
import pandas as pd
```

```python
train2=pd.read_csv('C:\\Users\\hp\\Documents\\DATA SCIENCE\\trial_projects\\train_modified.csv')
```

```python
test2=pd.read_csv('C:\\Users\\hp\\Documents\\DATA SCIENCE\\trial_projects\\test_modified.csv')
```

```python
#we divide the train2 dataset into independent X_train and remove the unnecessary variables
x_train2=train2.drop(['Item_Identifier','Outlet_Identifier','Item_Outlet_Sales'], axis=1)
#we create a dependent variable y_train2 which is our target variable
y_train2=train2.Item_Outlet_Sales
```

```python
x_test2=test2.drop(['Item_Identifier','Outlet_Identifier'], axis=1)
```

```python
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
```

```python
#create a train_test_split_model
xtrain,xtest,ytrain,ytest=model_selection.train_test_split(x_train2,y_train2,test_size=0.3, random_state=42)
```

```python
#initialize the linear model
lm=LinearRegression()
```

```python
#fit the training dataset into the model
lm.fit(xtrain,ytrain)
```

```python
print(lm.coef_)
```

```python
print(lm.intercept_)
```

```python
prediction=lm.predict(xtest)
```

```python
prediction
```

```python
#we validate the accuracy of the model by finding the mean squared error
from math import sqrt
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(ytest,prediction)
RMSE = math.sqrt(MSE)
```
